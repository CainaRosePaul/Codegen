#Load the dataset

```python
from datasets import Dataset,load_dataset

# Load the dataset using the load_dataset function
data = load_dataset("iamtarun/python_code_instructions_18k_alpaca")
```

#Formatting dataset

```python
def formatting_func(example):

    instruction = example['instruction'][0]
    output = example['output'][0]


    text = f"<start_of_turn>user\n{instruction}<end_of_turn> <start_of_turn>model\n{output}<end_of_turn>"

    return [text]
```

#Splitting

```python
from datasets import DatasetDict
subset_data = DatasetDict({"train": data['train'].select(range(200)).remove_columns(['input', 'prompt'])})
test_dataset = DatasetDict({"test": data['train'].select(range(300,400)).remove_columns(['input', 'prompt'])})
```

```python
print(subset_data)
```

```python
print(test_dataset)
```

```python
from codebleu import calc_codebleu
def generate_predictions_and_evaluate(model, tokenizer, test_dataset, max_length=200):
    predictions = []
    references = []

    for example in test_dataset['test']:
        input_text = example['instruction']


        chunks = [input_text[i:i+max_length] for i in range(0, len(input_text), max_length)]

        # Generate predictions for each chunk
        generated_texts = []
        for chunk in chunks:
            inputs = tokenizer(chunk, return_tensors="pt", truncation=True, padding=True)
            output_ids = model.generate(**inputs, max_length=max_length)
            generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
            generated_texts.append(generated_text)

        # Concatenate the generated chunks to form the complete sequence
        generated_text = ' '.join(generated_texts)

        reference_text = example['output']

        predictions.append(generated_text)
        references.append(reference_text)

    # Calculate CodeBLEU score
    codebleu_score = calc_codebleu(references, predictions, lang="python", tokenizer=None)

    return codebleu_score
```

#Model A

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import BitsAndBytesConfig
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,                     # Q = 4 bits
    bnb_4bit_use_double_quant=True,        # double quantization, quantizing the quantization constants for saving an additional 0.4 bits per parameter
    bnb_4bit_quant_type="nf4",             # 4-bit NormalFloat Quantization (optimal for normal weights; enforces w ∈ [-1,1])
    bnb_4bit_compute_dtype=torch.bfloat16  # Dequantize to 16-bits before computations (as in the paper)
)
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-2B-mono")
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-2B-mono", quantization_config=bnb_config, device_map="auto")
```

```python
print(model)
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

#Model A evaluation

```python

codebleu_score_model_A = generate_predictions_and_evaluate(model,tokenizer , test_dataset)
```
